```python
#Maybe change this to name of expense?

def totals(df, category, item=None):
    #Creates dataframe with only items that have the correct category
    df = df[df['Category']==category]
    display(df)
    total_cat = df['Price'].sum()
    print(f"This is the total for the category: ${total_cat}")
    return total_cat
    #This is ran if an item is specified
    if item != None:
        #Creates a dataframe with only that item 
        df2 =df[(df['Category']==category) & (df['Expense']==item)]
        #Gives the total for that item
        totals = pd.DataFrame({'Expense' : [item],'Total':[df2['Price'].sum()]})
        print("Displaying totals")
        display(totals)
def totals_value(df, category):
    df = df[df['Category']==category]
    total_cat = df['Price'].sum()
    return total_cat
    
```

```python
def difference(df_budget,df_actual):
    df = df_budget - df_actual
    display(df)
```

```python
def sep_list(inceptionList):
    list1 = []
    list2 = []
    list3 = []
    for i in inceptionList:
        list1.append(i[0])
        list2.append(i[1])
        list3.append(i[2])
    return list1, list2, list3
```
